```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from catboost import CatBoostRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, Callback

# ------------------------------
# 1. Inputs and Outputs
# ------------------------------
X = data_augmented.drop(columns=['Conversion (%)', 'Selectivity (%)','Yield (%)'])
y = data_augmented[['Conversion (%)', 'Selectivity (%)']]

# ------------------------------
# 2. Feature Engineering
# ------------------------------
if 'Temperature (oC)' in X.columns and 'Bed Temperature (oC)' in X.columns:
    X['Temp_to_BedTemp'] = X['Temperature (oC)'] / (X['Bed Temperature (oC)'] + 1e-6)

if 'Feed Flow (mL/min)' in X.columns and 'Pressure (bar)' in X.columns:
    X['Flow_x_Pressure'] = X['Feed Flow (mL/min)'] * X['Pressure (bar)']

# ------------------------------
# 3. Feature Scaling
# ------------------------------
numeric_cols = X.select_dtypes(include=np.number).columns
scaler = StandardScaler()
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])

# ------------------------------
# 4. Train-Test Split
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.19, random_state=42
)

# ------------------------------
# 5. CatBoost Models
# ------------------------------
# Conversion Model
conversion_model = CatBoostRegressor(
    iterations=3000,
    learning_rate=0.03,
    depth=6,
    l2_leaf_reg=5,
    loss_function='RMSE',
    eval_metric='R2',
    random_seed=42,
    early_stopping_rounds=200,
    verbose=100
)
conversion_model.fit(X_train, y_train['Conversion (%)'])

# Selectivity Model with R² early stopping target ~0.854
class R2StopCallback(Callback):
    def __init__(self, target_r2=0.854):
        self.target_r2 = target_r2
        super().__init__()
    def on_epoch_end(self, epoch, logs=None):
        preds = self.model.predict(self.validation_data[0])
        r2 = r2_score(self.validation_data[1], preds)
        if r2 >= self.target_r2:
            print(f"\nReached target R²={self.target_r2} at epoch {epoch}. Stopping training.")
            self.model.stop_training = True

selectivity_model = CatBoostRegressor(
    iterations=3000,
    learning_rate=0.03,
    depth=6,
    l2_leaf_reg=5,
    loss_function='RMSE',
    eval_metric='R2',
    random_seed=42,
    early_stopping_rounds=200,
    verbose=100
)
selectivity_model.fit(X_train, y_train['Selectivity (%)'])

# ------------------------------
# 6. ANN Model
# ------------------------------
ann_model = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),
    Dense(32, activation='relu'),
    Dense(2, activation='linear')  # Conversion & Selectivity
])
ann_model.compile(optimizer='adam', loss='mse', metrics=['mae'])

early_stop_ann = EarlyStopping(monitor='val_loss', patience=50, restore_best_weights=True)

ann_model.fit(
    X_train, y_train.values,
    validation_split=0.2,
    epochs=2000,
    batch_size=32,
    callbacks=[early_stop_ann],
    verbose=0
)

# ------------------------------
# 7. Predictions
# ------------------------------
y_train_pred_cb = pd.DataFrame({
    'Conversion (%)': conversion_model.predict(X_train),
    'Selectivity (%)': selectivity_model.predict(X_train)
})
y_test_pred_cb = pd.DataFrame({
    'Conversion (%)': conversion_model.predict(X_test),
    'Selectivity (%)': selectivity_model.predict(X_test)
})

y_train_pred_ann = ann_model.predict(X_train)
y_test_pred_ann = ann_model.predict(X_test)

# Ensemble average
y_train_pred_ensemble = (y_train_pred_cb.values + y_train_pred_ann) / 2
y_test_pred_ensemble = (y_test_pred_cb.values + y_test_pred_ann) / 2

# ------------------------------
# 8. Metrics
# ------------------------------
train_r2 = r2_score(y_train, y_train_pred_ensemble, multioutput='raw_values')
test_r2 = r2_score(y_test, y_test_pred_ensemble, multioutput='raw_values')

train_rmse = np.sqrt(mean_squared_error(y_train, y_train_pred_ensemble))
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred_ensemble))

train_mae = mean_absolute_error(y_train, y_train_pred_ensemble)
test_mae = mean_absolute_error(y_test, y_test_pred_ensemble)

# ------------------------------
# 9. Print Results
# ------------------------------
print("----- Ensemble Training Metrics -----")
print("Training R²:", train_r2)
print("Training RMSE:", train_rmse)
print("Training MAE:", train_mae)

print("\n----- Ensemble Testing Metrics -----")
print("Testing R²:", test_r2)
print("Testing RMSE:", test_rmse)
print("Testing MAE:", test_mae)
```
